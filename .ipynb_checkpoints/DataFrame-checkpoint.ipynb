# DATAFRAME MANIPULATION---PANDAS

#### A DataFrame is a two dimensional data structure where the data is arranged in a tabular format in rows and columns

### DataFrame creation

```python
import pandas as pd

# list of string
data_AnalystSkillSets=["EXcel","MYSQL","Python","Pandas","Matplotlib","Seaborn","PlotLib","Power Bi"]
df=pd.DataFrame(data_AnalystSkillSets)
print(df) 
```

```python
data_AnalystSkillSets=["EXcel","MYSQL","Python","Pandas","Matplotlib","Seaborn","PlotLib","Power Bi"]
df=pd.DataFrame(data_AnalystSkillSets,columns=["Skills_req"])
print(df) 
```

```python
Store_list=[["Vivo",200],["Oppo",300],["Samsung",100],["Apple",500]]
df=pd.DataFrame(Store_list,columns=["Store","Sales"])
print(df)
```

```python
# creating frame with dictionary

store_data={"products":["Vivo","Oppo","Samsung","Apple"],
            "sales":[200,300,800,1500]
           }
df=pd.DataFrame(store_data)
print(df)
# here keys works as columns
```

```python
# creating a dataframe with index from a dictionary
store_data={"products":["Vivo","Oppo","Samsung","Apple"],
            "sales":[200,300,800,1500]
           }
df=pd.DataFrame(store_data,index=["A","B","C","D"]) # passing a list of a index
print(df)
```

```python
# checking the datatype of each column/ key --> each column is a pandas series
print(type(df["products"]))
print(type(df["sales"]))
```

# Reading Data from Various Files

 #### Use the pandas read_csv() method to read the data from a csv file

```python
df_sales=pd.read_csv("xyz.csv")
print(df_sales)
```

#### • Use the pandas read_excel() method to read the data from xlsx file

```python
df_sales=pd.read_excel("xyz.xlsx")
print(df_sales)
```

#### If you have a csv file inside a zip file:

```python
import zipfile
with  zipfile.ZipFile("Xyz.zip") as z:
    with z.open("xyz.csv") as f:
        train=pd.read_csv(f,header=0)
        train.head() #here head() returns the first 5 records of the dataset
```

#### You can use the pandas read_csv() method to read the data from text file

```python
df_sales=pd.read_csv("xyz.txt",sep="\t")
print(df_sales)
```

#### • Use the pandas read_json() method to read the data from json file

```python
df_sales=pd.read_json("xyz.json")
print(df_sales)
```

#### • Use the pandas read_html() method to read the data from html file

```python
df_sales=pd.read_html("xyz.html")
print(df_sales)
```

# DataFrame Sorting

```python
df_empdata.sort_values("YearlyIncome").head(20) # herr head(20) means return first 20 data
```

```python
df_empdata.sort_values(["Sales","MonthlyIncome"],ascending=False).head(30)
```

```python
# • Get all the rows where Sales > 40000 and then sort all the rows by index using the sort_index() method
df_empdata[df_empdata.Sales>40000].sort_index(ascending=False).head(30)
```

# DataFrame Operations 

```python
# • Check the duplicate observations using the duplicated() method
# checking duplicate
# keep=False marks all duplicate rows as True
df_empdata.duplicated(keep=False)
```

#### • Find duplicate rows based on all columns • The parameter, keep=”first”, will select all duplicate rows except their 1st occurence

```python
# find the duplicates lets not consider First name and last name as duplicates
df_empdata=df_empdata.loc[:,(df_empdata.columns != "FirstName") & (df_empdata.columns != "LastName")]
df_empdata.head(50)
```

```python
df_dup_empdata=df_empdata[df_empdata.duplicated(keep="first")]
df_dup_empdata
```

## Drop Duplicates

```python
print(df_empdata.shape)

df_drop_duplicates_empdata=df_empdata.drop_duplicates()
print(df_drop_duplicates_empdata.shape)
print(df_empdata.shape)

```
